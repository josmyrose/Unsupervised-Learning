We do however need to transform categorical features into numeric ones where it makes sense to do so, as well as scaling the data due to distance being a key factor in clustering.

Here we use KMAENs clustering,because we calculate the distance in a cluster in order to do this ,we need numerical values instead of categorical value.

```python
X=pd.get_dummies(df)#Convert categorical variable into dummy/indicator variables. Data of which to get dummy indicators.
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Scalar_x=scaler.fit_transform(X)
```

```python
Scalar_x
```

```python
from sklearn.cluster import KMeans
```

Here we start no.of clusters in KMeans with 2,Because we don't know the correct number of clusters.In the marketing dataset,there is possibility of two groups. 

```python
model=KMeans(n_clusters=2)
```

```python
cluster_labels=model.fit_predict(Scalar_x)
```

```python
cluster_labels
```

```python
X['Cluster']=cluster_labels
```

```python
X.corr()['Cluster']
#correlation between Cluster group each varaible
```

```python
#To remove last column
X.corr()['Cluster'].iloc[:-1]
```

Choosing K Values

```python
ssd = []

for k in range(2,10):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(Scalar_x)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
```

#Sum of squared distances of samples to their closest cluster center, weighted by the sample weights if provided.

```python
plt.plot(range(2,10),ssd,'o--')
plt.xlabel("No.of K values")
plt.ylabel("Sum of squared distance")
```

From the above graph,we can decide no.of K values for the dataset.K value means that no.of clusters in the particular dataset.Here you can understand that if k=2,its sum of squared distance is 2.80.A good model is one with low inertia(Sum of Squared distance from each point to cluster center) AND a low number of clusters (K). However, this is a tradeoff because as K increases, inertia decreases.
